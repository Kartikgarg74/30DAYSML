```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Use all features of the Iris dataset
y = iris.target  # True labels (not used in clustering)

# Convert to a DataFrame for better understanding
df = pd.DataFrame(X, columns=iris.feature_names)

# Perform K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get the cluster centers and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Plotting the clustering results using the first two features
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.6, label="Data points")
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label="Centroids")
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('Feature 1 (Sepal Length)')
plt.ylabel('Feature 2 (Sepal Width)')
plt.legend()
plt.show()

# Display a table with original labels and predicted clusters
results = pd.DataFrame({'Original Labels': y, 'Cluster Labels': labels})
print(results.head())
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Use all features of the Iris dataset
y = iris.target  # True labels (not used in clustering)

# Convert to a DataFrame for better understanding
df = pd.DataFrame(X, columns=iris.feature_names)

# Define the model
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to the data
labels = dbscan.fit_predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.show()
# Display a table with original labels and predicted clusters
results = pd.DataFrame({'Original Labels': y, 'Cluster Labels': labels})
print(results.head())
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

iris = load_iris()
X = iris.data  # Use all features of the Iris dataset
y = iris.target  # True labels (not used in clustering)

# Convert to a DataFrame for better understanding
df = pd.DataFrame(X, columns=iris.feature_names)

# Define the model
hierarchical = AgglomerativeClustering(n_clusters=3)

# Fit the model to the data
labels = hierarchical.fit_predict(X)

# Plot the dendrogram
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()

# Display a table with original labels and predicted clusters
results = pd.DataFrame({'Original Labels': y, 'Cluster Labels': labels})
print(results.head())
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette score
score = silhouette_score(X, labels)
print("Silhouette Score:", score)
```

```python
from sklearn.metrics import davies_bouldin_score

# Calculate Davies-Bouldin index
db_index = davies_bouldin_score(X, labels)
print("Davies-Bouldin Index:", db_index)
```

## PCA

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load a sample dataset (Iris dataset as an example)
data = load_iris()
X = data['data']
y = data['target']
labels = data['target_names']

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a DataFrame for better handling and visualization
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Target'] = y
df_pca['Label'] = [labels[i] for i in y]

# Plot the results
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_pca,
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Label',
    palette='Set2',
    s=100,
    edgecolor='k'
)
plt.title("PCA Results: 2D Projection of Iris Dataset", fontsize=14)
plt.xlabel(f"Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}% variance explained)")
plt.ylabel(f"Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}% variance explained)")
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## t-SNE

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load a sample dataset (Iris dataset as an example)
data = load_iris()
X = data['data']
y = data['target']
labels = data['target_names']

# Perform t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
X_tsne = tsne.fit_transform(X)

# Create a DataFrame for better handling and visualization
df_tsne = pd.DataFrame(X_tsne, columns=['t-SNE Dimension 1', 't-SNE Dimension 2'])
df_tsne['Target'] = y
df_tsne['Label'] = [labels[i] for i in y]

# Plot the results
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_tsne,
    x='t-SNE Dimension 1',
    y='t-SNE Dimension 2',
    hue='Label',
    palette='Set2',
    s=100,
    edgecolor='k'
)
plt.title("t-SNE Results: 2D Projection of Iris Dataset", fontsize=14)
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## UMAP

```python
import umap.umap_ as umap
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load a sample dataset (Iris dataset as an example)
data = load_iris()
X = data['data']
y = data['target']
labels = data['target_names']

# Perform UMAP for dimensionality reduction
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)

# Create a DataFrame for better handling and visualization
df_umap = pd.DataFrame(X_umap, columns=['UMAP Dimension 1', 'UMAP Dimension 2'])
df_umap['Target'] = y
df_umap['Label'] = [labels[i] for i in y]

# Plot the results
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_umap,
    x='UMAP Dimension 1',
    y='UMAP Dimension 2',
    hue='Label',
    palette='Set2',
    s=100,
    edgecolor='k'
)
plt.title("UMAP Results: 2D Projection of Iris Dataset", fontsize=14)
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
